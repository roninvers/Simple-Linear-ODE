```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('euler_output_for_n_100.csv')
df['error'] = df['error'].abs()
```

```python
df.head()
```

```python
# Plot 1: Error term (y_euler - y_exact) vs t
plt.figure(figsize=(8, 5))
plt.plot(df['t'], df['error'], label='Absolute Error (y_euler - y_exact)', color='red')
plt.xlabel('t')
plt.ylabel('Absolute Error')
plt.title('Absolute Error between Euler Numerical Solution and Analytical Solution')
plt.grid(True)
plt.legend()
plt.show()


```

## Error Plot Description
The error plot shows the difference between the Euler numerical solution and the analytical solution over time. As expected, the error remains relatively small and may oscillate or grow slowly depending on the step size and the nature of the ODE. A smaller error indicates better accuracy of the Euler method. If we decrease the step size (increase \( n \)), the error typically decreases, demonstrating the convergence of the numerical method to the exact solution.


```python
# Plot 2: y_euler and y_exact vs t on the same plot
plt.figure(figsize=(8, 5))
plt.plot(df['t'], df['y_euler'], label='Euler Numerical Solution', color='blue')
plt.plot(df['t'], df['y_exact'], label='Analytical Solution', color='green', linestyle='dashed')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Euler Numerical Solution and Analytical Solution vs Time')
plt.grid(True) # Helps to visualize better for smaller interval size.
plt.legend()
plt.show()
```

## Euler Numerical Solution and Analytical Solution Plot Description
This plot displays both the Euler numerical solution and the analytical solution for the differential equation as functions of time. The analytical solution serves as the exact reference, while the Euler solution provides an approximation. The closer the Euler curve is to the analytical curve, the more accurate the numerical method. Deviations between the two curves highlight the numerical error introduced by the discretization. Now since the interval is split into 1000 parts we see the error is very less and the two solutions almost overlap.
