```python
import numpy as np

# Define the function for the right-hand side of the ODE: y' = cos(t) - y
def f(t, y):
    return np.cos(t) - y

# Euler's method function for any given n
def euler_method(n):
    t_initial = 0      # Start of the interval
    t_final = 5        # End of the interval
    y_initial = 1      # Initial value of y at t = 0
    h = (t_final - t_initial) / n  # Step size
    t = np.linspace(t_initial, t_final, n+1)  # Time points
    y = np.zeros(n+1)                         # Array to store y values
    y[0] = y_initial                          # Set initial condition
    for i in range(n):
        y[i+1] = y[i] + f(t[i], y[i]) * h     # Euler update
    return t, y

# Example usage:
t, y = euler_method(20)
# Note that, you can change the input to 100 or even 1000 and have the plot to say how it behaves
```

```python
import matplotlib.pyplot as plt

# Plot the results for n = 20
plt.figure(figsize=(8, 5))
plt.plot(t, y, 'o-', label="Euler's Method (n=20)")
plt.xlabel('t')
plt.ylabel('y')
plt.title("Euler's Method for $y' = \cos(t) - y$, $y(0)=1$")
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Analytical solution as a function
def analytical_solution(t):
    return 0.5 * np.sin(t) + 0.5 * np.cos(t) + 0.5 * np.exp(-t)
```

```python
# Compute for n = 100
n = 100
t, y_euler = euler_method(n)
y_exact = analytical_solution(t)
error = np.abs(y_euler - y_exact) # Compute the absolute difference between the Euler and analytical solutions at each time point
```

```python
# Plot the error with n = 100
plt.figure(figsize=(10, 4))
plt.plot(t, error, label='Absolute Error (n=100)')
plt.xlabel('t')
plt.ylabel('Absolute Error')
plt.title('Error between Euler Method and Analytical Solution (n=100)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute for n = 1000
n = 1000
t, y_euler = euler_method(n)
y_exact = analytical_solution(t)
error = np.abs(y_euler - y_exact) # Compute the absolute difference between the Euler and analytical solutions at each time point
```
